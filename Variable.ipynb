# Variable (O'zgaruvchi)

### O‘zgaruvchi (variable) nima?


**O‘zgaruvchi** — bu kompyuter xotirasida ma’lum bir ma’lumot (qiymat)ni saqlash uchun ajratilgan joy. Soddaroq qilib aytganda, o‘zgaruvchini biror ma’lumotni saqlash uchun ishlatiladigan quti yoki konteyner sifatida tasavvur qilish mumkin. Bu qutining ichidagi narsa — ma’lumot (qiymat) bo‘lib, qutining o‘zi — o‘zgaruvchi hisoblanadi.

#### O'zgaruvchilarni yaratish va ularga qiymat berish


Python-da o'zgaruvchi yaratish uchun, uning nomini yozib qiymat berish kifoya,boshqa tillardagi ka'bi o'zgaruvchilarni avvaldan e'lon qilish yo'q.


Quyidagi misolda, `oquv_markaz` bu o'zgaruvchining nomi `BEU Academy` esa o'zgaruvchining qiymati.

```python
oquv_markaz = "BEU Academy"
```

```python
ism = "Tom"
yosh = 10
print(ism)
print(yosh)
```

Yuqoridagi misolda biz `ism` va `yosh` degan o'zgaruvchilarni yaratdik va `ism`ga `Tom` `yosh`ga esa `10` qiymatlarini berdik

### O'zgaruvchining qiymatini o'zgartirish

O'zgaruvchi (`variable`) deb atalishining sababi, uning qiymati istalgan vaqt o'zgartirish mumkinligi:

```python
ism = "Tom"
print(ism)
ism = "Jerry"
print(ism)
```

Yuqoridagi misolda `ism` nomli o'zgaruvchiga avval `Tom` degan qiymatni bergan edik keyin uinig qiymatini `Tom`dan `Jerry`ga o'zgartirdik.

## O'ZGARUVCHILARNI NOMLASH

O'zgaruvchilarga nom berishda quyidagi qoidalarga amal qiling:

- O'zgaruvchi nomi harf yoki pastki chiziq (`_`) bilan boshlanishi kerak
- O'zgaruvchi nomi raqam bilan boshlanishi mumkin emas
- O'zgaruvchi nomida faqatgina lotin alifbosi harflari (`A-z`), raqamlar (`0-9`) va pastki chiziq (`_`) qatnashishi mumkin
- O'zgaruvchi nomida bo'shliq (пробел) bo'lishi mumkin emas
- O'zgaruvchi nomida katta-kichik harflar turlicha talqin qilinadi (`ism`, `ISM`, va `Ism` uchta turli o'zgaruvchi)
  
Qo'shimcha qoida sifatida:

- O'zgaruvchi nomini kichik harflar bilan yozing.
- O'zgaruvchi nomida 2 va undan ortiq so'z qatnashsa ularning orasini pastki chiziq (`_`) bilan ajrating (`ism_sharif="Falonchi Pistonchiev`")
- O'zgaruvchiga tushunarli nom bering (`y=20` emas `yosh=20`, `d="Korea"` emas `davlat = "Korea"` va hokazo)
- Shuningdek o'zgaruvchilarga Pythonda ishlatiladigan funktsiyalar va maxsus kalit so'zlarning (`keywords`) nomini bermang. Kalit so'zlar ro'yhatini -- ko'rish uchun Spyder konsolida avval `help()` deb yozing va Enter tugmasini bosing. Keyin esa `keywords` deb kiritib, yana Enter bosing. Marhamat, ekraningizda Pythondagi maxsus kalit so'zlar ro'yhatini ko'ryapsiz:

```python
help()
```
